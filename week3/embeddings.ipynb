# Embeddings : For better contexual Understanding


```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
result =embedding.embed_query("AI is the future for humanity")
print(str(result))
```

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv 
load_dotenv()
```

```python
embedding2 = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
result = embedding2.embed_query("AI is the future for humanity")
print(str(result))
```

```python
### processing list of documents all together for embedding

docs = [
    "budapest is the capital of hungary",
    "paris is the capital of france",
    "berlin is the capital of germany",
    "madrid is the capital of spain"    
]
```

```python
result = embedding.embed_documents(docs) 
print(str(result))
```
